```python
import time
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# ──────────────────────────────────────────────────────────────────────────────
# 1) Prepare data
#    • df_ddos: your DataFrame with all numeric features + 'label_binary'
# ──────────────────────────────────────────────────────────────────────────────
X = df_ddos.drop(columns="label_binary")
y = df_ddos["label_binary"]

# Scale once for those models that need it
scaler = StandardScaler().fit(X)
X_scaled = scaler.transform(X)

# ──────────────────────────────────────────────────────────────────────────────
# 2) Define models
#    tuple = (name, estimator, use_scaled_flag)
# ──────────────────────────────────────────────────────────────────────────────
model_defs = [
    ("Logistic L1",     LogisticRegression(penalty='l1', solver='liblinear', max_iter=1000), True),
    ("Random Forest",   RandomForestClassifier(random_state=42),                          False),
    ("XGBoost‑CPU",     XGBClassifier(eval_metric='logloss',random_state=42),False),
    ("SVM‑RBF",         SVC(kernel='rbf', probability=True, random_state=42),             True),
    ("KNN",             KNeighborsClassifier(),                                         False),
    ("GaussianNB",      GaussianNB(),                                                    False),
]

# ──────────────────────────────────────────────────────────────────────────────
# 3) Cross‑validation setup
# ──────────────────────────────────────────────────────────────────────────────
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = []

print("Starting 5‑fold CV on each model…")
for name, model, use_scaled in model_defs:
    print(f"\n→ {name}:")
    X_in = X_scaled if use_scaled else X.values

    t0 = time.time()
    cv_res = cross_validate(
        model,
        X_in, y,
        cv=cv,
        scoring='accuracy',
        n_jobs=-1,
        return_train_score=False
    )
    elapsed = time.time() - t0

    mean_acc = np.mean(cv_res['test_score'])
    print(f"   • Completed in {elapsed:.2f}s; CV accuracy = {mean_acc:.3f}")

    results.append({
        "Model":          name,
        "CV Time (s)":    round(elapsed, 3),
        "CV Accuracy":    round(mean_acc,  3)
    })

print("\nAll models finished.\n")

# ──────────────────────────────────────────────────────────────────────────────
# 4) Assemble and display results
# ──────────────────────────────────────────────────────────────────────────────
cv_results = pd.DataFrame(results).set_index("Model")
print(cv_results)
```
